```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_dataset = sns.load_dataset("titanic")
titanic_dataset
```

```python
sns.barplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```

```python
sns.barplot(x = "survived", y = "sex", data = titanic_dataset)
plt.show()
```

```python
sns.barplot(x = "sex", y = "survived", data = titanic_dataset, color='blue')
plt.show()
```

```python
sns.barplot(x = "sex", y = "survived", data = titanic_dataset, palette = 'hls')
plt.show()
```

```python
sns.barplot(x = "class", y = "survived", hue = "embark_town", data = titanic_dataset)
plt.show()
```

```python
sns.barplot(x = "class", y = "survived", hue = "embark_town", errorbar = None, data = titanic_dataset)
plt.show()
```

```python
sns.barplot(x = "class", y = "survived", hue = "embark_town", order = ["Third", "Second", "First"], data = titanic_dataset)
plt.show()
```
