```python
image_path = '/content/drive/MyDrive/Colab Notebooks/Computer_vision/people.jpg'
```

```python
img = cv2.imread(image_path)
```

```python
img.shape
```

```python
from google.colab.patches import cv2_imshow
gray_image=  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
gray_image.shape
```

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor = 1.05, minNeighbors=5, minSize = (40,40)
)

```

#minNeighbors (e.g., 5): This parameter controls how many neighboring detections are required for a region to be classified as a face. Higher values reduce false positives (non-face regions being detected as faces) but may also cause you to miss some actual faces. minsize it represents the minimum size (width and height) of the detected face

```python
face
```

```python
for ( x, y, w, h) in face:
  cv2.rectangle(img, (x,y), (x + w, y + h ), (0,255,0), 4)
```
