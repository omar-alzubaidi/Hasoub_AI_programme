```python
import pandas as pd

df = pd.read_csv("car_malfunctions.csv")

df.info()
```

```python
xpoints = df['Age']
y1points = df['malfunctions']
y2points = df['DaysOff']

plt.subplot(1,2,1)

plt.scatter(xpoints, y1points)

plt.xlabel("Age")
plt.ylabel("Malfunctions")


plt.subplot(1,2,2)

plt.scatter(xpoints, y2points)

plt.xlabel("Age")
plt.ylabel("DaysOff")

plt.show()
```
