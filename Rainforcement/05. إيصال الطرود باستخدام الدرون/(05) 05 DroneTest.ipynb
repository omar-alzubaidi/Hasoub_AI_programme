```python
import pickle
def load_q_table_and_env(q_table_file, env_file):
    # تحميل مصفوفة القيم
    with open(q_table_file, 'rb') as f:
        q_table = pickle.load(f)
    # تحميل البيئة
    with open(env_file, 'rb') as f:
        env = pickle.load(f)
    return q_table, env
```

```python
# مسار مجلد العمل
working_folder='/content/drive/MyDrive/RLModels/'
# مسار ملف مصفوفة القيم
q_table_Drone_path = working_folder + "q_table_Drone.pkl"
# مسار ملف البيئة
env_pkl_Drone_path = working_folder + "env_pkl_Drone.pkl"
```

```python
# تحميل مصفوفة القيم والبيئة
q_table, env = load_q_table_and_env(q_table_Drone_path, env_pkl_Drone_path)
```
