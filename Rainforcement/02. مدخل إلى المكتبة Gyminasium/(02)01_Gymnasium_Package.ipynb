```python
# تهيئة البيئة مع تحديد نمط العرض
env = gym.make('FrozenLake-v1', render_mode='ansi')
state, info = env.reset()
grid=env.render()
print(grid)
```

```python
# عدد الحالات
n_states = env.observation_space.n
print(n_states)
```

```python
# عدد الأفعال
n_actions = env.action_space.n
print(n_actions)
```

```python
# التهيئة والعرض
env = gym.make('FrozenLake-v1', render_mode='rgb_array', is_slippery=False)
# تحديد عدد الخطوات الأعظمي
from gymnasium.wrappers import TimeLimit
env = TimeLimit(env, max_episode_steps=20)
# إعادة التهيئة
state, info = env.reset()
# العرض
grid=env.render()
grid
```

```python
# التحرك للأسفل
next_state, reward, terminated, truncated, info = env.step(1)
print(next_state)
print(reward)
print(terminated)
print(truncated)
grid=env.render()
grid
```

```python
# إعلاق البيئة
env.close()
```
