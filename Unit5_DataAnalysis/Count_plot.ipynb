```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x ='sex', data = df)
plt.show()
```

```python
sns.countplot(x ='sex', data = df, order = ['Female', 'Male'])
plt.show()
```

```python
sns.countplot(data=df, x='day')
plt.show()
```

```python
sns.countplot(data=df, x='day', order=['Sat', 'Sun', 'Thur', 'Fri'])
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts().index)
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts(ascending=True).index)
plt.show()
```

```python
sns.countplot(x ='sex', hue = "smoker", data = df)
plt.show()
```

```python
ax = sns.countplot(x ='sex', hue = "smoker", data = df)

for label in ax.containers:
    ax.bar_label(label)
    
plt.show()
```

```python
sns.countplot(y ='sex', hue = "smoker", data = df)
plt.show()
```

```python
sns.countplot(x ='sex', palette = "Set2", data = df)
plt.show()
```
