```python
import gymnasium as gym
env = gym.make("CarRacing-v3", render_mode="rgb_array")
```

```python
# رفع الدرايف
from google.colab import drive
drive.mount('/content/drive')
```

```python
# مسار مجلد النماذج
working_folder='/content/drive/MyDrive/RLModels/'
#model_file_name="HF_ppo_cart_racing"
# عدد دورات التدريب
total_timesteps=1000000
model_file_name=f"ppo_car_racing_{total_timesteps}"
# مسار ملف النموذج المدرب
model_file_path = working_folder + model_file_name
```

```python
#!pip install shimmy>=0.2.1
```

```python
# مكتبة التعلم المُعزز
from stable_baselines3 import PPO
# تحميل النموذج
model = PPO.load(model_file_path)
```

```python
# مكتبات التسجيل
from gymnasium.wrappers import RecordEpisodeStatistics, RecordVideo

# مسار مجلد الفيديويات
video_folder= '/content/drive/MyDrive/CarRacingVideos/'

# عدد الحلقات
num_eval_episodes = 5

# تسجيل الفيديو
env = RecordVideo(env, video_folder=video_folder, name_prefix=f"trained_{total_timesteps}",
                  episode_trigger=lambda x: True)
# تسجيل الإحصائيات
env = RecordEpisodeStatistics(env, buffer_length=num_eval_episodes)

# الدوران في حلقات
for episode_num in range(num_eval_episodes):
    state, info = env.reset()
    # مؤشر الخروح من الحلقة
    done = False
    while not done:
      # استخدام النموذج المدرب للتنبؤ بالفعل المناسب
      action,_= model.predict(state, deterministic=True)
      # الانتقال للحالة التالية
      state, reward, terminated, truncated, info = env.step(action)
      # تحديث مؤشر الخروح من الحلقة
      if terminated or truncated:
          done = True

# طباعة الإحصائيات
print(f'Episode time taken: {env.time_queue}')
print(f'Episode total rewards: {env.return_queue}')
print(f'Episode lengths: {env.length_queue}')

# إغلاق البيئة
env.close()
```
