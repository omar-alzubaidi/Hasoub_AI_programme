```python
model_name = 'nlptown/bert-base-multilingual-uncased-sentiment'
```

```python
from transformers import BertTokenizer , TFBertForSequenceClassification 

tokenizer = BertTokenizer.from_pretrained(model_name)

model = TFBertForSequenceClassification.from_pretrained(model_name)
```

```python
text = "I like my job"
inputs = tokenizer(text)
print(inputs)
input_ids = inputs['input_ids']
predictions = model.predict([input_ids])
logits = predictions.logits
import numpy as np
predicted_class = np.argmax(logits)
predicted_class
```

# The code tokenizes a sentence ("I like my job") into numbers.
# The model predicts sentiment based on these numbers.
# The highest probability class is selected as the final prediction.


```python
sentiment_maping = {
    0: 'very negative',
    1: 'negative',
    2: 'neutral',
    3: 'positive',
    4: 'very positive',
}
predicted_sentiment = sentiment_maping[predicted_class]

print("The text is predicted to have a sentiment:{}".format(predicted_sentiment))

    
```

```python
import pandas as pd

csv_file = 'IMDB Dataset.csv'

df = pd.read_csv(csv_file)

df
```

```python
firist_ten_rows = df.head(10)
firist_ten_rows
```

```python
import numpy as np

sentiment_mapping = {
    0: 'Very negative',
    1: 'negative',
    2: 'neutral',
    3: 'positive',
    4: 'very positive'
}

firist_ten_rows_copy = firist_ten_rows.copy()  

predicted_sentiment = [] 

for title in firist_ten_rows_copy['review']:
    inputs = tokenizer(title, return_tensors="tf")
    input_ids = inputs['input_ids']

    outputs = model(input_ids)
    logits = outputs.logits

    predicted_class = np.argmax(logits, axis=1).flatten()[0]

    sentiment = sentiment_mapping[predicted_class]
    predicted_sentiment.append(sentiment)         

    print("predicted id: ", predicted_class)
    print("predicted_sentiment:", sentiment)

firist_ten_rows_copy['predicted_sentiment'] = predicted_sentiment

firist_ten_rows_copy
```
