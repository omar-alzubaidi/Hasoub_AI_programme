```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "dreamlike-art/dreamlike-photoreal-2.0"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
prompts = [
    "Majestic mountain landscape at sunset, vibrant colors, serene atmosphere",
    "Futuristic city skyline at night, neon lights, cyberpunk vibe",
    "Calm beach scene with gentle waves, sunrise, warm tones, minimalistic"
]
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
image.save('/content/drive/MyDrive/Colab Notebooks/Computer_vision/'+ f'picture_{i}.jpg')
```
