```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'column_a':[1, 2, 3, 4],
    'column_b':['a', 'b', 'c', 'd'],
    'column_c':[True, True, False, True]
})
```

```python
df2 = pd.DataFrame({
    'column_a':[1, 2, 9, 10],
    'column_b':['a', 'k', 'l', 'm'],
    'column_c':[False, False, False, True]
})
```

```python
df = pd.concat([df1, df2])
df
```

```python
df = pd.concat([df1, df2], keys=['df1', 'df2'])
df
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df = pd.concat([df1, df2], axis=1)
df
```
