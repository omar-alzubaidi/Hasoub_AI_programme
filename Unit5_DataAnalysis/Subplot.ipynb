```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car_malfunctions.csv")

y1 = df['Age']
y2 = df['malfunctions']

plt.subplot(1, 2, 1)
plt.plot(y1, y2, marker = 'o')

plt.subplot(1, 2, 2)
plt.plot(y2, y1, marker = 'o')

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(y1, y2, marker = 'o')

plt.subplot(2, 1, 2)
plt.plot(y2, y1, marker = 'o')

plt.show()
```

```python
y1 = df['Age']
y2 = df['malfunctions']
y3 = df['DaysOff']

plt.subplot(1, 3, 1)
plt.plot(y1, y2, marker = 'o')

plt.subplot(1, 3, 2)
plt.plot(y2, y1, marker = 'o')

plt.subplot(1, 3, 3)
plt.plot(y2, y3, marker = 'o')

plt.show()
```
